### Python Operators


Operators are used to perform operations on variables and values.

In the example below, we use the + operator to add together two values:

```python
print(5 + 2)
```

###### Python divides the operators in the following groups:
<ul>
	<li>Arithmetic operators</li>
	<li>Assignment operators</li>
	<li>Comparison operators</li>
	<li>Logical operators</li>
	<li>Identity operators</li>
	<li>Membership operators</li>
	<li>Bitwise operators</li>
</ul>

   
#### 1. Arithmetic Operators


Arithmetic operators are used with numeric values to perform common mathematical operations:



<ul>
<li>Addition</li>
<li>Subtraction</li>
<li>Multiplication</li>
<li>Division</li>
<li>Modulus</li>
<li>Exponent</li>
</ul>

##### Addition and Subtraction

```python
x= 5   #  variable name => x and value => 5
y= 3 + 2 # y = 5
z= x + y  # z = 5 + 5 = 10
print(z)
print(x + 2)
```

```python
length= +2 # unary operator + mean positive value   
negativeNum= -5 
print ("length: "+ str(length))
print ("negative: "+ str(negativeNum))
```

##### Multiplication and Division

```python
x= 2
y= 5
z= x* y
print(z)
```

```python
x= 6
y= 3
print(x/y)
print(x//y) # Always use floor if we have 2.6 it will be 2 
```

##### Modulus

```python
x = 5
y = 2

print(x % y)
```

##### Exponent

```python
x = 2
y = 3

print(x ** y) #same as 2*2*2
```

### 2.Comparison Operators


Comparison operators are used to compare two values:



```python
x = 5
y = 3

print(x == y)

# returns False because 5 is not equal to 3
```

```python
x = 5
y = 3

print(x != y)

# returns True because 5 is not equal to 3
```

```python
x = 5
y = 3

print(x > y)

# returns True because 5 is greater than 3
```

```python
x = 5
y = 3

print(x <= y)

# returns False because 5 is neither less than or equal to 3
```

### 3.Logical Operators or Bitwise Operators


Logical operators are used to combine conditional statements:



```python
x = 5

print(x > 3 and x < 10)

# returns True because 5 is greater than 3 AND 5 is less than 10
```

```python
x = 5

print(x > 3 or x < 4)

# returns True because one of the conditions are true (5 is greater than 3, but 5 is not less than 4)
```

```python
x = 5

print(not(x > 3 and x < 10))

# returns False because not is used to reverse the result
```
